###  함수의 결과로 외부에 두 개 이상의 값을 전달할 수 있다

```python
def comp(x, y):
    sum = x+y
    mul = x*y
    return sum, mul # 두 개 이상의 값 리턴 가능 (다른 언어에서는 불가능하다)

a = 3
b = 4
c, d = comp(a, b) # 두 개 이상의 값을 한번에 두 개의 변수에 저장(tuple의 특성)
print('Sum is', c)
print('Multiplication is', d)

```

### 숫자 출력에 관한 예제

```python
sum = 0
d = 1.5
while d < 3.5:
    print(d)
    sum += d
    d += 0.1
print("합 = ", sum)
```

```python
sum = 0
d = 1.5
while d < 3.5:
    print(f'{d:6.2f}')
    sum += d
    d += 0.1
print(f"합 = {sum:6.2f}") #= print(f"합 = ", f"{sum:6.2f}")
```

### 가운데 정렬, 왼쪽 정렬, 오른쪽 정렬

```python
name = 'han'
print(f'My name is {name: ^10}.') #가운데 정렬
print(f'My name is {name: <10}.') # 왼쪽 정렬
print(f'My name is {name: >10}.') # 오른쪽 정렬
```

```python
a = 123
print(f'a={a:10}') # 기본은 오른쪽 정렬이다
print(f'a={a:>10}') # 오른쪽 정렬
print(f'a={a:^10}') # 가운데 정렬
print(f'a={a:<10}') # 왼쪽 정렬
```

### List, Tuple, Set, Dictionary

### Set

* 중복된 원소를 제거
* {}로 표시
* empty set은 set()로 정의
* integer, float, tuple, string을 원소로 가질 수 있다.

```python
a = {1, 3, 3, 5, 7}
print(a)
```

```python
a = set()
print(a)
a.add(9)
print(a)
```

```python
a = {1, 2.3, "ABC", (5, 6, "D")} # type((5, 6, "D")) = tuple
print(a)
```

* mutable(변할 수 있는) element인 list, dictionary를 원소로 가질 수 없다.

```python
a = {[1, 2, 3]} # 리스트[1, 2, 3]을 원소로 한다.
print(a)
```

```python
a = set([1, 2, 3]) #리스트 [1, 2, 3]이 갖고 있는 값들을 원소로 한다.
print(a)
```

```python
a = set()
print(type(a))
```

* set_name.add()은 하나의 원소를 추가한다.

```python
a = {1, 2, 3}
print(a)
a.add(4)
print(a)
```

```python
a.add(5, 6)
```

* set_name.update(): 괄호 내에는 tuple, list, string 또는 set.

```python
a = {1, 2, 3}
a.update([4, 5])
print(a)
```

```python
a = {1, 2, 3}
a.update((4, 5))
print(a)
```

```python
a = {(1, 2), (3, 4)}
print(a)
a.update((5, 6))
print(a)
```

```python
a = {(1, 2), (3, 4)}
print(a)
a.add((5, 6))
print(a)
```

* set_name.discard() 또는 set_name.remove(): 괄호 내에는 원소

```python
a = {1, 2, 3}
a.discard(1)
print(a)
a.remove(2)
print(a)
```

* discard는 set내에 원소가 없도 가능. remove는 원소가 없을 경우 error

```python
a = {1, 2, 3}
a.discard(4)
print(a)
a.remove(4)
print(a)
```

* 여러 개의 원소를 제거

```python
a = {1, 2, 3, 4, 5}
b = set()
for x in a:
    if x not in {3, 5}:
        b.add(x)
print(b)
```

```python
a = {1, 2, 3, 4, 5}
a = {x for x in a if x not in {3, 5}}
print(a)
```

* set operation - union

```python
a = {1, 2, 3}
b = {3, 4, 5}
c = a|b
print(c)

print(a.union(b))
```

* set operation - intersection

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a & b)

print(a.intersection(b))
```

* set operation - difference

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a - b)

print(a.difference(b))
```

* set operation - symmetric difference

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a ^ b)
print(a.symmetric_difference(b))
```

* set operation - member

```python
a = {1, 2, 3}
print(2 in a)
print(5 in a)
```

### Dictionary

* {'key' : 'value'}
* {}로 표시
* key는 immutable{변할 수 없는} type을 가져야 한다.
* string, number, tuple 가능

```python
a = {}
print(a)

a = {1: 'car', 2: 'house'}
print(a)
```

* value로는 다양한 형태가 가능

```python
a = {'kk': 'car', 2: [5, 7]}
print(a)
```

* dict()는 dictionary로 만든다.

```python
a = {'college': 'eni', 'dept': 'ce'}
print(a)

a = dict({'college': 'eni', 'dept': 'ce'})
print(a)

a = dict([('college', 'eni'), ('dept', 'ce')]) # 리스트 안의 두개의 튜플을 각각 딕셔너리로 만든다
print(a)
```

* key를 이용하여 value를 찾는 방법

```python
a = {'college': 'eni', 'dept': 'ce'}
print(a['college'])
print(a['dept'])
print(a.get('dept'))
```

* key의 value를 변경하는 방법

```python
a = {'college': 'eni', 'dept': 'ce'}
a['id'] = 4321 # key 와 value 추가
print(a)
```

* membership 확인하는 방법

```python
a = {'college': 'eni', 'id': 1234, 'age': 23}
print('id' in a)
print('home' in a)
```

* iterating 방법

```python
a = {'college': 'eni', 'id':1234, 'age': 23}
for i in a:
    print(a[i])
```

## 리스트는 mutable(값을 바꿀수)있다.

* b=a는 리스트 a를 b라고 나타낼 수 있다.
* c=a[:]는 리스트a를 리스트c에 복사한다.

```python
a = [1, 2, 3, 4]
b = a
c = a[:]
print(a, b, c)
a.append(5)
print(a, b, c)
```
